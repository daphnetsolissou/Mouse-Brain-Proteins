```python
def find_unique_common(data1,data2,field,fields):
    # belong to CC_data_2dge and doesn't belong to CC_data_HRMS
    data1_unique = data1[~data1[field].isin(data2 [field])]
    print(len(data1_unique))
    # belong to CC_data_HRMS and doesn't belong to data_sharma
    data2_unique = data2[~data2[field].isin(data1 [field])]
    print(len(data2_unique))
    common = pd.merge(data1, data2, on=fields)
    print(len(common))
    total_prot = pd.concat([data1_unique,data2_unique,common])
    print(len(total_prot))
    return total_prot
```

```python
import re
import requests
import urllib

def word_break(text_br):
    words = re.findall(r'\b\S+\b', text_br) 
    return words

def retrieve_prot_inf(query):
    global K
    K=K+1
    print (query,"  ---  ", K, end="\r")
    ac_uniprot = ""
    id_uniprot = ""
    res = urllib.request.urlopen("http://www.uniprot.org/uniprot/" + query + ".txt").read()
    s = str(res, 'utf-8')
#     print(s)
    lines = s.splitlines()
#     print(lines)
#   print(res)
    cellular_comp = [] #C
    molecular_func = [] #F
    biological_proc = [] #P
    cellular_comp_id = [] #C
    molecular_func_id = [] #F
    biological_proc_id = [] #P
    
    for i in range(len(lines)):
        match_word=word_break(lines[i])
        if len(lines[i])>2:
            if match_word[0]=="DR" and match_word[1]=="GO":
#                 print(lines[i])
#                 print(match_word)
#                 print(len(match_word))
#                 print(len(match_word)-5)
#                 print("-------------------------------")
                gene_id = match_word[2]
                
                if match_word[3][0] == "C":
#                     print(match_word[3][2:])
                    sentence = match_word[3][2:]
                    #because we must exclude DR GO GO:342552 and the first and last word
                    for j in range(4,len(match_word)-1):
                        sentence = sentence +" "+ match_word[j]
                    cellular_comp.append(sentence)
                    cellular_comp_id.append(gene_id)
                elif match_word[3][0] == "F":
#                     print(match_word)
#                     print(match_word[3][2:])
                    sentence = match_word[3][2:]
                    #because we must exclude DR GO GO:342552 and the first and last word
                    for j in range(4,len(match_word)-1):
                        sentence = sentence +" "+ match_word[j]
                    molecular_func.append(sentence)
                    molecular_func_id.append(gene_id)
                elif match_word[3][0] == "P":
#                     print(match_word[3][2:])
                    sentence = match_word[3][2:]
                    #because we must exclude DR GO GO:342552 and the first and last word
                    for j in range(4,len(match_word)-1):
                        sentence = sentence +" "+ match_word[j]
                    biological_proc.append(sentence)
                    biological_proc_id.append(gene_id)
                
    return cellular_comp,molecular_func,biological_proc,cellular_comp_id,molecular_func_id,biological_proc_id

```

```python
import os
from goatools.obo_parser import GODag

# Load a small test GO DAG and all the optional relationships,
# like 'regulates' and 'part_of'
godag = GODag('go-basic.obo',
              optional_attrs={'relationship'})
```

```python
from goatools.gosubdag.gosubdag import GoSubDag
from goatools.godag.go_tasks import get_go2parents

 


def find_ancestors(GO_ID):
    optional_relationships = set()
    go2parents_isa = get_go2parents(godag, optional_relationships)
    ancestor_1 = ""
    ancestor_2 = ""
    ancestor_3 = ""
    try:
        gosubdag_r0 = GoSubDag([GO_ID], godag, prt=None)
        ancestors = list(gosubdag_r0.rcntobj.go2ancestors[GO_ID])
#         print(ancestors)
        levels = []
        for n in gosubdag_r0.rcntobj.go2ancestors[GO_ID]:
            levels.append(godag[n].level)
#         print(levels)   

    #     print(levels.index(0))
        ancestor_1 = ancestors[levels.index(0)]
        if 2 in levels:
            ancestor_3 = ancestors[levels.index(2)]
            ancestor_2 = ancestors[levels.index(1)]
        elif 1 in levels:
            ancestor_2 = ancestors[levels.index(1)]
            ancestor_3 = GO_ID
        else:
            ancestor_2 = GO_ID
            ancestor_3 = ""    
        return ancestor_1,ancestor_2,ancestor_3
    except KeyError as e:
        print ('I got a KeyError - reason' + str(e))
        return ancestor_1,ancestor_2,ancestor_3
        
print(find_ancestors("GO:0060352"))
```
