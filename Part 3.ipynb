```python
from goatools.gosubdag.gosubdag import GoSubDag
from goatools.godag.go_tasks import get_go2parents

 


def find_ancestors(GO_ID):
    optional_relationships = set()
    go2parents_isa = get_go2parents(godag, optional_relationships)

    
    not_root = True
    while(not_root):
        gosubdag_r0 = GoSubDag([GO_ID], godag, prt=None)
#         print(list(gosubdag_r0.rcntobj.go2ancestors[GO_ID]))
        if len(list(gosubdag_r0.rcntobj.go2ancestors[GO_ID]))==1:
            ancestor_1 = go2parents_isa[GO_ID]
            return (list(ancestor_1)[0],GO_ID,"")
            not_root = False 
        if len(list(gosubdag_r0.rcntobj.go2ancestors[GO_ID]))==2:
            ancestor_2 = go2parents_isa[GO_ID]
            ancestor_1 = go2parents_isa[list(ancestor_2)[0]]
            return (list(ancestor_1)[0],list(ancestor_2)[0],GO_ID)
            not_root = False
        if len(list(gosubdag_r0.rcntobj.go2ancestors[GO_ID]))==3:
            ancestor_3 = go2parents_isa[GO_ID]
#             print(ancestor_3)
            ancestor_2 = go2parents_isa[list(ancestor_3)[0]]
#             print(ancestor_2)
            ancestor_1 = go2parents_isa[list(ancestor_2)[0]]
            return (list(ancestor_1)[0],list(ancestor_2)[0],list(ancestor_3)[0])
            not_root = False
        else:
            GO_ID = list(go2parents_isa[GO_ID])[0]
        
print(find_ancestors("GO:0003714"))
```

```python
from goatools.gosubdag.gosubdag import GoSubDag
from goatools.godag.go_tasks import get_go2parents

 


def find_ancestors1(GO_ID):
    optional_relationships = set()
    go2parents_isa = get_go2parents(godag, optional_relationships)
    ancestor_1 = ""
    ancestor_2 = ""
    ancestor_3 = ""
    try:
        gosubdag_r0 = GoSubDag([GO_ID], godag, prt=None)
        ancestors = list(gosubdag_r0.rcntobj.go2ancestors[GO_ID])
#         print(ancestors)
        levels = []
        for n in gosubdag_r0.rcntobj.go2ancestors[GO_ID]:
            levels.append(godag[n].level)
#         print(levels)   

    #     print(levels.index(0))
        ancestor_1 = ancestors[levels.index(0)]
        if 2 in levels:
            ancestor_3 = ancestors[levels.index(2)]
            ancestor_2 = ancestors[levels.index(1)]
        elif 1 in levels:
            ancestor_2 = ancestors[levels.index(1)]
            ancestor_3 = GO_ID
        else:
            ancestor_2 = GO_ID
            ancestor_3 = ""    
        return ancestor_1,ancestor_2,ancestor_3
    except KeyError as e:
        print ('I got a KeyError - reason' + str(e))
        return ancestor_1,ancestor_2,ancestor_3
        
print(find_ancestors1("GO:0060352"))
```

```python

# Choose a deep leaf-level GO ID associated with "bacteria"
DESC = 'nucleus'            # GO Term name contains this
NSPC = 'cellular_component'  # Desired namespace

# Create a chooser function which returns True or False
def chooser(goterm):
    """Choose a leaf-level GO term based on its name"""
    b_match = DESC in goterm.name
    # True if GO term is leaf-level (has no children)
    b_leaf = not goterm.children
    # True if GO term is in 'cellular_component' namespace (nspc)
    b_nspc = goterm.namespace == NSPC
    return b_match and b_leaf and b_nspc

# Get GO terms with desired name in desired GO DAG branch
go_ids_selected = set(o.item_id for o in godag.values() if chooser(o))
print(godag.values())
print('{N} {desc} GO terms'.format(N=len(go_ids_selected), desc=DESC))
```
